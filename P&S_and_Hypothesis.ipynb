6> Use a chi-square test to examine the relationship between categorical variables, such as state and order status

```python
H0 = "State and Order Status are independent"
H1 = "State and Order Status are dependent"

ana_table = pd.crosstab(all_data["State"],all_data["Order_Status"])
print(ana_table)
_,p,_,_ = stats.chi2_contingency(ana_table)
print(p)
alpha = 0.05

if p < alpha:
    print(H1)
else:
    print(H0)
```

7> Conduct an ANOVA to determine if there are significant differences in the average order amount across different states or cities

```python
H0 = "There are significant differences in the average order amount across different states or cities"
H1 = "There are no significant differences in the average order amount across different states or cities"

df_guj = df[df["state"] == "GUJARAT"]["amount_paid"]
df_har = df[df["state"] == "HARYANA"]["amount_paid"]
df_jar = df[df["state"] == "JHARKHAND"]["amount_paid"]
df_karn = df[df["state"] == "KARNATAKA"]["amount_paid"]
df_mah = df[df["state"] == "MAHARASHTRA"]["amount_paid"]
df_ker = df[df["state"] == "KERALA"]["amount_paid"]
df_on = df[df["state"] == "ON"]["amount_paid"]
df_pradesh = df[df["state"] == "PRADESH"]["amount_paid"]


_,p = stats.f_oneway(df_guj,df_har,df_jar,df_karn,df_mah,df_ker,df_on,df_pradesh)
print(p)
alpha = 0.05

if p < alpha:
    print(H1)
else:
    print(H0)
```

8> Analyze trends in the order data over time by plotting orderd_date against amount

```python
df['month_year'] = df['Order_Date'].str.split('-').apply(lambda x:x[0]) + ' - ' +  df['Order_Date'].str.split('-').apply(lambda x:x[1])
month_wise = df.groupby(by=df['month_year'])
month_wise_sum = month_wise["amount_paid"].aggregate((sum))
month_wise_sum = pd.DataFrame({'month':month_wise_sum.index,'total amount':month_wise_sum.values})
sns.lineplot(month_wise_sum,x=month_wise_sum['month'],y=month_wise_sum['total amount'])
plt.xticks(rotation=90)
```

10> Perform Recency, Frequency, Monetary (RFM) analysis to segment customers based on their purchasing behavior.

```python
# Measures how recently a customer has made a purchase.
# Measures how often a customer makes a purchase within a specific time period.

full_data['Order_Date'] = pd.to_datetime(full_data['Order_Date'],format='%d-%m-%Y')

analysis_date = full_data['Order_Date'].max()
print(f"Analysis date is:{analysis_date}")

# Calculate RFM metrics
rfm_df = full_data.groupby('Customer_Name').agg({
    'Order_Date': lambda x: (analysis_date - x.max()).days,
    'Order_Number': 'count',
    'Amount_Paid': 'sum'
}).reset_index()

rfm_df.columns = ['Customer_Name', 'Recency', 'Frequency', 'Monetary']

# Display the RFM dataframe
print(rfm_df.head())

# full_data
```
